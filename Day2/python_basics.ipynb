```python
import pandas as pd
# Create a simple DataFrame
data = {
    'Student': ['A', 'B', 'C', 'D'],
    'Math': [85, 90, 78, 88],
    'Science': [92, 85, 80, 95]
}
df = pd.DataFrame(data)
print(df)
# Basic statistics
print("\\nAverage Marks:")
print(df[['Math', 'Science']].mean())
```
