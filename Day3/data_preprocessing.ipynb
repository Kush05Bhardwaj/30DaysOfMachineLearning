```python
## üì• 1. Import Libraries
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
## üßæ 2. Load Example Dataset
```

```python
# Example dataset with missing and categorical values
data = {
'Age': [25, 30, np.nan, 22, 28],
'Salary': [50000, 54000, 58000, np.nan, 62000],
'City': ['Delhi', 'Mumbai', 'Delhi', 'Bangalore', np.nan],
'Purchased': ['Yes', 'No', 'Yes', 'No', 'Yes']
}

df = pd.DataFrame(data)
print("Original Data:")
print(df)
```

```python
## üîß 3. Handle Missing Values
```

```python
# Fill numerical columns with mean
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())

# Fill categorical column with mode
df['City'] = df['City'].fillna(df['City'].mode()[0])
print("After Handling Missing Values:")
print(df)
```

```python
## üè∑Ô∏è 4. Encode Categorical Variables
```

```python
# Encode 'City' using LabelEncoder
le_city = LabelEncoder()
df['City'] = le_city.fit_transform(df['City'])

# Encode 'Purchased'
df['Purchased'] = df['Purchased'].map({'Yes': 1, 'No': 0})

print("Encoded Data:")
print(df)
```

```python
## ‚öñÔ∏è 5. Feature Scaling
```

```python
# Standardize numerical columns\n",
scaler = StandardScaler()
df[['Age', 'Salary']] = scaler.fit_transform(df[['Age', 'Salary']])
print("Scaled Data:")
print(df)
```

```python
## ‚úÇÔ∏è 6. Train-Test Split
```

```python
X = df[['Age', 'Salary', 'City']]
y = df['Purchased']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print("Training Set:")
print(X_train)
print("Test Set:")
print(X_test)
```
