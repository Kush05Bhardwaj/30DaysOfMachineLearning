```python
import numpy as np

# Define two vectors
a = np.array([2, 3])
b = np.array([4, 1])
```

```python
# Basic operations
add = a + b
sub = a - b
dot = np.dot(a, b)

print("a + b =", add)
print("a - b =", sub)
print("Dot Product =", dot)
```

```python

# Define matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[2, 0], [1, 2]])
```

```python
# Operations
C = np.dot(A, B)
det_A = np.linalg.det(A)
inv_A = np.linalg.inv(A)

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("A x B =\n", C)
print("Determinant of A =", det_A)
print("Inverse of A:\n", inv_A)
```

```python
A = np.array([[2, 1], [1, 1]])
b = np.array([5, 3])

# Solve for x
x = np.linalg.solve(A, b)

print("Solution (x, y) =", x)
```
