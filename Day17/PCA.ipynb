```python
## 1. Import Libraries & Load Dataset
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load dataset
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
## 2. Perform PCA (2 Components)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

```python
## 3. Plot PCA Components
plt.figure(figsize=(7,5))
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap='viridis', s=50)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Visualization (2 Components)")
plt.colorbar(label='Target Class')
plt.show()
```

```python
## 4. PCA + KMeans (Bonus Visualization)
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto')
clusters = kmeans.fit_predict(X_pca)

plt.figure(figsize=(7,5))
plt.scatter(X_pca[:,0], X_pca[:,1], c=clusters, cmap='turbo', s=50)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Clustering on PCA-Reduced Space")
plt.show()
```

```python
## 5. Compare Original vs PCA Variance
plt.figure(figsize=(6,4))
plt.bar(['PC1','PC2'], pca.explained_variance_ratio_, color='skyblue')
plt.title("Explained Variance Ratio")
plt.ylabel("Variance")
plt.show()
```
