```python
## üßÆ 1. Generate Synthetic Data
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Generate synthetic data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

plt.scatter(X, y, color='blue', label='Data points')
plt.title('Generated Dataset')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
## ‚öôÔ∏è 2. Linear Regression ‚Äî From Scratch
```

```python
# Calculate m and c manually (Closed-form solution)
X_b = np.c_[np.ones((100, 1)), X]  # add bias term
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

print(f"Intercept (c): {theta_best[0][0]:.2f}")
print(f"Slope (m): {theta_best[1][0]:.2f}")
```

```python
# Predictions
y_pred_scratch = X_b.dot(theta_best)
```

```python
# Plot
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred_scratch, color='red', label='From Scratch Line')
plt.title('Linear Regression (From Scratch)')
plt.legend()
plt.show()
```

```python
## ü§ñ 3. Linear Regression ‚Äî Using Scikit-learn
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
y_pred_sklearn = model.predict(X)

print(f"Intercept (c): {model.intercept_[0]:.2f}")
print(f"Slope (m): {model.coef_[0][0]:.2f}")
print(f"R¬≤ Score: {r2_score(y, y_pred_sklearn):.4f}")
```

```python
# Visualization
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred_sklearn, color='green', label='Sklearn Line')
plt.legend()
plt.title('Linear Regression (Scikit-learn)')
plt.show()
```
