```python
## 1. Import Libraries & Train Simple Model
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import pickle
import joblib
```

```python
# Load Dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Model
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)

print("Model trained successfully!")
```

```python
## 2. Save Model Using Pickle
filename_pickle = "logistic_model.pkl"

with open(filename_pickle, "wb") as f:
    pickle.dump(model, f)

filename_pickle
```

```python
## 3. Load Model Using Pickle
with open(filename_pickle, "rb") as f:
    loaded_pickle_model = pickle.load(f)

sample = [[5.1, 3.5, 1.4, 0.2]]
pred = loaded_pickle_model.predict(sample)[0]
iris.target_names[pred]
```

```python
## 4. Save Model Using Joblib
filename_joblib = "logistic_model.joblib"

joblib.dump(model, filename_joblib)

filename_joblib
```

```python
## 5. Load Model Using Joblib
loaded_joblib_model = joblib.load(filename_joblib)
pred2 = loaded_joblib_model.predict(sample)[0]
iris.target_names[pred2]
```
