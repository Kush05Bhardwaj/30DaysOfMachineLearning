```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Sample dataset
data = [12, 15, 14, 10, 18, 20, 14, 13, 15]
```

```python
# Calculations
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data, keepdims=True)[0][0]
variance = np.var(data)
std_dev = np.std(data)
```

```python
print(f"Mean: {mean:.2f}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
```

```python
# Visualization
plt.figure(figsize=(6,4))
plt.hist(data, bins=6, color='skyblue', edgecolor='black')
plt.axvline(mean, color='red', linestyle='--', label=f'Mean = {mean:.2f}')
plt.axvline(median, color='green', linestyle='--', label=f'Median = {median}')
plt.title('Histogram of Data')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
import random
```

```python
# Simulate 100 coin tosses
events = ['Heads', 'Tails']
outcomes = [random.choice(events) for _ in range(100)]
prob_heads = outcomes.count('Heads') / len(outcomes)
prob_tails = outcomes.count('Tails') / len(outcomes)
```

```python
print(f"P(Heads): {prob_heads:.2f}")
print(f"P(Tails): {prob_tails:.2f}")
```

```python
# Visualize outcomes
plt.bar(['Heads', 'Tails'], [prob_heads, prob_tails], color=['#66b3ff', '#ff9999'])
plt.title('Coin Toss Probability Distribution')
plt.ylabel('Probability')
plt.show()
```
