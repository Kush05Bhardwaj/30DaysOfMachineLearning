```python
## üì• 1. Import Libraries & Load Dataset
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Features: {feature_names}")
print(f"Target Names: {target_names}")
```

```python
## ‚öôÔ∏è 2. Train Decision Tree Model
```

```python
# Train Decision Tree Classifier
clf = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=42)
clf.fit(X_train, y_train)
```

```python
# Predictions
y_pred = clf.predict(X_test)
```

```python
# Evaluation
acc = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {acc:.2f}")
```

```python
## üå≤ 3. Visualize the Decision Tree
```

```python
plt.figure(figsize=(14,8))
plot_tree(clf, feature_names=feature_names, class_names=target_names, filled=True, rounded=True, fontsize=10)
plt.title("Decision Tree Visualization")
plt.show()
```

```python
## üìä 4. Confusion Matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
disp.plot(cmap='Greens')
plt.title('Confusion Matrix')
plt.show()
```

```python
## üìà 5. Feature Importance
```

```python
importance = pd.Series(clf.feature_importances_, index=feature_names)
importance.sort_values(ascending=True).plot(kind='barh', color='teal')
plt.title('Feature Importance')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.show()
```
