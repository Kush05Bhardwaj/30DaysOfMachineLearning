```python
## 1. Import Libraries & Load Dataset
```

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load dataset
iris = load_iris()
X = iris.data[:, :2]  # use first 2 features for visualization
```

```python
## 2. Elbow Method to Find Best K
```

```python
inertia_list = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(X)
    inertia_list.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(6,4))
plt.plot(K_range, inertia_list, marker='o')
plt.title("Elbow Method - Best K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia (SSE)")
plt.grid(True)
plt.show()
```

```python
## 3. Train K-Means with Optimal K (K=3)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto')
clusters = kmeans.fit_predict(X)
centroids = kmeans.cluster_centers_

print("Cluster Centers:")
print(centroids)
```

```python
## 4. Visualize Clusters & Centroids
```

```python
plt.figure(figsize=(6,4))
plt.scatter(X[:,0], X[:,1], c=clusters, cmap='viridis', s=50)
plt.scatter(centroids[:,0], centroids[:,1], color='red', s=200, marker='X')
plt.title("K-Means Clusters (K=3)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
## 5. Predict New Sample Cluster
```

```python
sample = np.array([[5.0, 3.4]])
pred_cluster = kmeans.predict(sample)[0]
print(f"Predicted Cluster for {sample}: {pred_cluster}")
```
